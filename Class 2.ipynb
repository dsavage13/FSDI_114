Definitions:
a.) A stack is a linear data structure that follows (LIFO), meaning that the last element to be added is     the first to be removed. Two real world examples are email inbox's, and photo gallerys on a device.
b.) A queue is also a linear data structure except it follows First in first out (FIFO). A couple of real world examples are: Email outboxes, or even like a car was queue. 

Primary Operations:
Stack:
Push: Adds an element to the top of the stack.
Pop: Removes and returns the top element of the stack.
Peek: Returns the top element of the stack without removing it.
IsEmpty: Checks if the stack is empty.
Size: Returns the number of elements in the stack.

Queue: 
Enqueue: Adds an element to the end (rear) of the queue.
Dequeue: Removes and returns the front element of the queue.
Front: Returns the front element of the queue without removing it.
IsEmpty: Checks if the queue is empty.
Size: Returns the number of elements in the queue.

Key Differences: 
The first and main difference is the Last In First Out(LIFO) vs First In First Out (FIFO) principles that they follow. Another key difference is reference points, in a stack you just need to reference the top, while in a queue you need to reference both front and rear. Some examples are like the email inbox vs outbox. When you have your inbox, the last message to be recieved is the first one to be displayed. But in the outbox, (or sent box) the first message in gets sent first.
